```python
import numpy as np
from scipy.stats import entropy
from math import log, e

def entropy2(labels, base=None):
  

    n_labels = len(labels)

    if n_labels <= 1:
        return 0

    value,counts = np.unique(labels, return_counts=True)
    probs = counts / n_labels
    
    n_classes = np.count_nonzero(probs)
    #print(probs[1])

    if n_classes <= 1:
        return 0

    ent = 0.

    # Compute entropy
    base = e if base is None else base
    #for i in probs:
    ent -= probs[1] * log(probs[1], base)
    return ent

    
```

```python
import random

def min_entropy(label, start, end):
    #print(f'buscando {objetivo} entre {lista[comienzo]} y {lista[final - 1]}')
    if start > end:
        return False

    half = (start + end)// 2
    entropy = entropy2(label)
    entropy_search =  entropy2(label[:start])+entropy2(label[start:])

    if entropy_search < entropy:
        return True
    elif entropy2(label[:half]) > entropy:
        return (min_entropy(label, half + 1, end), end)
    else:
        return (min_entropy(label, start, half - 1), start
   


```

```python
def L(label, x):
    return entropy2(label[:x])+entropy2(label[x:])
```

```python
%%time
import time


L_func =[]
index = []

for i in range(10,len(df),10000):
    L_T = L(yy_sorted, i)
    L_func.append(L_T)
    index.append(i)
```

```python
results_sorted =pd.DataFrame({'index': index, 
                        'L_func': L_func})
results_sorted.describe()
```
