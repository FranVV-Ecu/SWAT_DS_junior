## 1. Assignment problem.

The attached file (Assignment_problem.csv) contains a cost matrix for tasks execution. There are
20 tasks (T1, T2... T20) and 10 workers (W1, W2... W10). It is necessary to distribute tasks among
workers so that the total cost is minimal and the following conditions must be met:

- Each task has to be assigned
- All employees should have at least one task, but not more than 3.

As an answer, please write the total obtained cost and a vector of length 10, each element of
which corresponds to the number of tasks assigned to the corresponding worker.
For example, 395, [3,1,2,1,2,3,3,1,1,3].

```python
import pandas as pd
import numpy as np
```

```python
path_file = r"C:\Users\Asus\Desktop\SWAT mobility\Assignment_problem.csv"
data = pd.read_csv(path_file, index_col = 0)
#Checking out data
data
```

```python
from collections import Counter
#Getting dataframe to compute
df =data.copy()
l =[]
total = 0
lenght = int(len(df))

#Assigment one minimal cost task to every worker
for _ in range(len(df)):
    
    value = df.min().min()
    total += value
    indx = df.stack().idxmin()
    
    df = df.drop(indx[0])
    df = df.drop(indx[1], axis =1 )
    l.append(indx[0])
    
    
task = data[df.columns]    #Getting only the left tasks not assigned

#Assigment tasks of minimal cost  
for _ in range(len(df.columns) - len(df)): 
    val = task.min().min()
    total += val
    ind = task.stack().idxmin()
    task = task.drop(ind[1], axis =1 )
    l.append(ind[0])
    c = Counter(l)
    if c.get(ind[0]) < 3:
        continue
    else:
        task = task.drop(ind[0])
#Get object with workers and task quantity, and total value fo cost task 
c, total
```

```python
from collections import OrderedDict
#Sorting  dict with workers by key and assigned tasks
mydict = dict(c)
sort_dict = OrderedDict(sorted(mydict.items()))
```

```python
#Vector of assigned task 
vector_tasks = [v for k, v in sort_dict.items()]

#Get total cost and vector assigned tasks by worker from W1 to W10 in ascending order
total, vector_tasks
```
