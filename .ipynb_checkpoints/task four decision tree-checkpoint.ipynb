```python
import pydotplus 
from sklearn.tree import export_graphviz
from sklearn import tree
def tree_graph_to_png(tree, feature_names, png_file_to_save):
    tree_str = export_graphviz(tree, feature_names=feature_names, 
                                     filled=True, out_file=None)
    graph = pydotplus.graph_from_dot_data(tree_str)  
    graph.write_png(png_file_to_save)
```

```python
from sklearn.tree import DecisionTreeClassifier

x_tree = DecisionTreeClassifier(criterion='entropy',max_depth = 3, min_samples_leaf=100, random_state=17)
x_tree.fit(df['x'].values.reshape(-1, 1), df['y'].values)

```

```python
import matplotlib.pyplot as plt

fn=['x']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(x_tree,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```

```python
params_dist = dict(df['x'].describe())
params_dist
```

```python
lower_bound = int(params_dist['std']-params_dist['mean'])
higher_bound = int(params_dist['std']+params_dist['mean'])
```

```python
df_f = df[df.x.between(lower_bound, higher_bound)]
```

```python
df_f.info()
```

```python
from math import log
def entropy( vector):
    n1 = vector.value_counts()[1]
    nt = int(len(vector))
    return -(n1/nt)*log((n1/nt),2)
```

```python
f_tree = DecisionTreeClassifier(criterion='entropy',max_depth = 1, min_samples_leaf=10, random_state=85)
f_tree.fit(df_f['x'].values.reshape(-1, 1), df_f['y'].values)
```

```python
import matplotlib.pyplot as plt

fn=['x']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(f_tree ,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```
