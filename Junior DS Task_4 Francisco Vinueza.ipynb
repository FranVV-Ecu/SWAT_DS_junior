The attached file (Entropy.csv) contains huge table (500 mln. records) with 2 columns: x and y. Column y contains only 0 and 1. Let‚Äôs define entropy of the table T as:


$ùêº(ùëá) = ‚àí {\frac {ùëÅ_{1}}{ùëÅ_{T}}}‚àó ùëôog_{2}({\frac {ùëÅ_{1}}{ùëÅ_{T}}})$,
$T_{x \leq x^{*} }$


where N1 is the number of ‚Äú1‚Äù in column y and NT is the total number of records in table. We split this table by two separate tables: $T_{x \leq x^{*} }$ and $T_{x > x^{*} }$. $T_{x \leq x^{*} }$ contains all records from T with ${x \leq x^{*}}$. You need to find ùë•‚àó which minimizes obj = ùêº($T_{x \leq x^{*} }$) + ùêº($T_{x > x^{*} }$). The number of record of each sub-table must greater than 10.

As a result, please provide ùë•‚àó, obj value and the code.

```python
import pandas as pd
import numpy as np
```

```python
#Checking out data 
file = r"C:\Users\Asus\Desktop\SWAT mobility\Entropy.csv"
df = pd.read_csv(file)
df.sample(10)
```

## Aproach 1

In this aproach every record is consider as an independent event, in this case the order of data frame is not changed and split iterating by i index, in two subsets. $T_{x \leq x^{*}_{i} }$ and $T_{x > x^{*}_{i} }$, and compute entropy of i subset . As a result we obtain a Series obj_L

```python
from math import log

# Entropy function 

def ent(x,y):
    if x== 0:
        ent = 0
    else:
        ent = -(x/y)*log((x/y),2)
    return ent
```

```python
import numpy as np
# In order to split data frame calculate N1 and Nt fom cumulative summ in two ways. 

df['n1'] = df.y.cumsum() # N1 number of 1 in ascending acumulation

df['nt'] = df.index #Nt number of recors of every value x in acumulaive sum

df['inv_nt'] = df.nt.values[::-1] # N1 number of 1 in an inverted acumulation

df['inv_n1'] = df.n1.values[::-1] # N1 number of records in an inverted acumulation
```

```python
# Calulate entropy from every row that gaves us a split frame in two parts.

df['ent_a'] = np.vectorize(ent)(df.n1, df.nt).round(6)

df['ent_b'] = np.vectorize(ent)(df.inv_n1, df.inv_nt).round(6)
```

```python
#Compute function L as a sum of subsests splited on value i
df['L'] = df['ent_a'] + df['ent_b']
df
```

```python
#Aplly boundery of minimal number of records in every subset 10
obj = df.loc[ 10: int(len(df)-10) , ['x', 'L']] # resulting dataset of L value in i subset 
obj.sample(5)
```

```python
value_x = obj[obj.L == obj.L.min()] # Minimal value of function obj
value_x
```

# Aproach 2

In this aproach we consider that the event x are dependent and indexed by a unique number that is value x, hence consider a lable x value as an unitary subset. From this point of view in order to reduce dataframe it used method groupby and sorted by value.

```python
#Sorting by 'x'. Pivot data frame by summ of ones 'y_sum' and number of records in every value 'y_count' 
data = pd.read_csv(file)
df_group =  data.groupby('x').agg(['sum','count']).reset_index()
df_group.columns = ['_'.join(col).strip() for col in df_group.columns.values]
df_group.head()
```

```python
import numpy as np
# In order to split data frame calculate N1 and Nt fom cumulative summ in two ways. 

df_group['n1'] = df_group.y_sum.cumsum() # N1 number of 1 in ascending acumulation

df_group['nt']= df_group.y_count.cumsum() #Nt number of recors of every value x in acumulaive sum

df_group['inv_n1'] = df_group[::-1].y_sum.cumsum() # N1 number of 1 in inverse  acumulation

df_group['inv_nt'] = df_group[::-1].y_count.cumsum() #Nt number of t in inverse acumulation
```

```python
# Calulate entropy from every row that gaves us a split frame in two parts.

df_group['ent_a'] = np.vectorize(ent)(df_group.n1, df_group.nt).round(6)

df_group['ent_b'] = np.vectorize(ent)(df_group.inv_n1, df_group.inv_nt).round(6)
```

```python
# Summ of two dataframe at every value f x 
df_group['L_obj'] = df_group.ent_a + df_group.ent_b

# Cheking out results of dataframe gouped
df_group.sample(10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, axes = plt.subplots( figsize=(12, 5))
plt.scatter(df_group.x_, df_group.ent_a, label = 'Entropy subset A', s=1) 
plt.scatter(df_group.x_, df_group.ent_b, label = 'Entropy subset B', s=1) 
plt.scatter(df_group.x_, df_group.L_obj, label = 'Entropy obj L ', s=1) 
plt.ylim(0.25,0.98)
plt.xlabel('Values x')
plt.ylabel('Entropy ')
plt.legend()
plt.xlim(-30000,229000)
plt.grid()
fig.tight_layout()
plt.show()
```

```python
obj_L = df_group.loc[ 10: int(len(df_group)-10) , ['x_', 'L_obj']]
obj_L.sample(5)
```

```python
value_obj = obj_L[obj_L.L_obj == obj_L.L_obj.min()]
value_obj
```
